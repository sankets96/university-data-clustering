```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
```

```python
new = df.drop('private',axis=1).copy()

```

```python
new
```

```python
model = KMeans()
```

```python
k_val = KElbowVisualizer(model,k=(1,12)).fit(new)
```

```python
#number of cluster is 4
final_model = KMeans(n_clusters=4,init="k-means++",max_iter=100,random_state=0).fit(new)
```

```python
final_model.labels_
```

```python
final_model.inertia_
```

```python
final_model.n_iter_
```

```python
final_model.n_clusters
```

```python
def converter(private):
    if private == "Yes":
        return 1
    else:
        return 0
```

```python
df["Cluster"] = df['private'].apply(converter)
```

```python

df.head()
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(df["Cluster"], final_model.labels_))
print(classification_report(df["Cluster"],final_model.labels_))
```
